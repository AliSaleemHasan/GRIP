## import dependencies

```python
import easyocr # used library to extract text from images
import requests # to get image from url 
import cv2 # to get image from webcam 
from matplotlib import pyplot as plt # to output image
import math
%matplotlib inline
```

## define draw function to draw bounding boxes on output image

```python
def draw_output(image,output):
    '''
    Draw bounding boxes around detected words with corresponding word above them
    '''

    # check if there is output, if not function will be aborted
    if len(output)== 0 :
        plt.imshow(image)
        return

    # loop through output bounding boxes and words to draw each one separatly 
    for box in output:
        
        # get bounding box points as integers 
        points = [[int(point) for point in boxPoints] for boxPoints in box[0]]

        # we just need top left and bottom right points in order to draw the corresponding rectangle
        topLeft = tuple(points[0])
        bottomRight = tuple(points[2])

        # draw detection rectangles
        cv2.rectangle(image,topLeft,bottomRight,(0,255,255),math.ceil(box[2] * 2))

    
    plt.imshow(image)
```

## apply text extraction to image and draw output on it 

```python
def applayOCR(imagePath,lang=['en']):
    image = cv2.imread(imagePath)
    reader = easyocr.Reader(['en'])
    output = reader.readtext(imagePath)
    draw_output(image,output)
    # print output 
    [print(text[3]) for text in output]
```
